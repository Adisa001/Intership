```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
header_texts = [tag.text.strip() for tag in header_tags]
header_dict = {f'Header_{i+1}': header_texts[i] for i in range(len(header_texts))}
df = pd.DataFrame(list(header_dict.items()), columns=['Tag', 'Content'])
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.imdb.com/chart/top/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
movie_list = soup.find('tbody', {'class': 'lister-list'})
names = []
ratings = []
years = []
for movie in movie_list.find_all('tr')[1:51]:
    name = movie.find('td', {'class': 'titleColumn'}).find('a').text
    rating = float(movie.find('td', {'class': 'ratingColumn imdbRating'}).find('strong').text)
    year = int(movie.find('td', {'class': 'titleColumn'}).find('span', {'class': 'secondaryInfo'}).text.strip('()'))
    
    names.append(name)
    ratings.append(rating)
    years.append(year)
df = pd.DataFrame({'Name': names, 'Rating': ratings, 'Year of Release': years}) 
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.imdb.com/list/ls063315661/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
movies = []
for item in soup.find_all(class_='lister-item-content')[1:51]:
    title = item.find(class_='lister-item-header').a.text
    rating = float(item.find(class_='ratings-imdb-rating').strong.text)
    year = int(item.find(class_='lister-item-year').text.strip('()'))
    movies.append({'Title': title, 'Rating': rating, 'Year': year})
movies_df = pd.DataFrame(movies)
print(movies_df)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 

url = 'https://presidentofindia.nic.in/former-presidents.htm'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
president_list = soup.find_all('table', class_='tablepress tablepress-id-33')
names = []
terms = []

for president in president_list[1:]:
    name = president.find_all('td', class_='column-1')
    term = president.find_all('td', class_='column-2')
    
    for n, t in zip(name, term):
        names.append(n.text.strip())
        terms.append(t.text.strip())

df = pd.DataFrame({'Name': names, 'Term of office': terms})
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.ice-cricket.com/rankings/mens/odi"


response = requests.get(url)


soup = BeautifulSoup(response.content, "html.parser")

table = soup.find("table", {"class": "table table-responsive"})


rankings = []
for row in table.find_all("tr")[1:]:
    cells = row.find_all("td")
    rank = cells[0].text.strip()
    team = cells[1].text.strip()
    matches = cells[2].text.strip()
    points = cells[3].text.strip()
    rating = cells[4].text.strip()
    rankings.append([rank, team, matches, points, rating])


df = pd.DataFrame(rankings, columns=["Rank", "Team", "Matches", "Points", "Rating"])


df = df.head(10)


print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.ice-cricket.com/rankings/womens-odi/batting/'


response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


players = []
teams = []
ratings = []

for i in range(1, 11):
    player = soup.find('div', {'class': 'rankings-table__player-name', 'rank': str(i)})
    team = soup.find('div', {'class': 'rankings-table__team-name', 'rank': str(i)})
    rating = soup.find('div', {'class': 'rankings-table__rating', 'rank': str(i)})
    if player and team and rating:
        players.append(player.text.strip())
        teams.append(team.text.strip())
        ratings.append(rating.text.strip())


data = {'Player': players, 'Team': teams, 'Rating': ratings}


df = pd.DataFrame(data)


print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


headline_news = []
headline_links = []
headline_sources = []

articles = soup.find_all('div', class_='Card-title')
for article in articles:
    title = article.find('a').text
    link = article.find('a')['href']
    source = article.find('span', class_='Card-source').text
    headline_news.append(title)
    headline_links.append(link)
    headline_sources.append(source)


df = pd.DataFrame({'Title': headline_news, 'Link': headline_links, 'Source': headline_sources})


print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)


soup = BeautifulSoup(response.content, "html.parser")


articles_list = soup.find_all("li", class_="pod-listing-item")


authors_list = []
for article in articles_list:
    authors = []
    for author in article.find_all("span", class_="author"):
        authors.append(author.text.strip())
    authors_list.append(", ".join(authors))


data = pd.DataFrame({"Authors": authors_list})


print(data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.direct.co.in/restaurant-details" 
response = requests.get(url)


soup = BeautifulSoup(response.content, "html.parser")


restaurant_elements = soup.find_all("div", class_="restaurant-item")


restaurant_names = []
restaurant_addresses = []
restaurant_cuisines = []


for restaurant_element in restaurant_elements:
    name = restaurant_element.find("h3").text.strip()
    address = restaurant_element.find("p", class_="address").text.strip()
    cuisine = restaurant_element.find("p", class_="cuisine").text.strip()

    
    restaurant_names.append(name)
    restaurant_addresses.append(address)
    restaurant_cuisines.append(cuisine)


data = pd.DataFrame({
    "Restaurant Name": restaurant_names,
    "Restaurant Address": restaurant_addresses,
    "Restaurant Cuisine": restaurant_cuisines
})


print(data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.dineout.co.in/delhi-restaurants/'


response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


restaurant_names = []
restaurant_ratings = []


restaurant_elements = soup.find_all('div', class_='restnt-card clearfix')


for restaurant_element in restaurant_elements:
    
    name = restaurant_element.find('h2', class_='restnt-name').text.strip()
    restaurant_names.append(name)
    
    
    rating = restaurant_element.find('span', class_='rating-val').text.strip()
    restaurant_ratings.append(rating)


df = pd.DataFrame({'Restaurant Name': restaurant_names, 'Rating': restaurant_ratings})


print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.dineout.co.in/delhi-restaurants/'


response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


restaurant_names = []
restaurant_image_urls = []


restaurant_elements = soup.find_all('div', class_='restnt-card clearfix')


for restaurant_element in restaurant_elements:
    
    name = restaurant_element.find('h2', class_='restnt-name').text.strip()
    restaurant_names.append(name)
    
    
    image_url = restaurant_element.find('img')['data-src']
    restaurant_image_urls.append(image_url)


df = pd.DataFrame({'Restaurant Name': restaurant_names, 'Image URL': restaurant_image_urls})


print(df)
```
