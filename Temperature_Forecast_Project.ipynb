```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/Dataset2/main/temperature.csv')
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dropna(inplace=True)



```

```python
x = df.drop(columns=['Next_Tmin', 'Next_Tmax', 'Date'])
y_t_min = df['Next_Tmin']
y_t_max = df['Next_Tmax']

x_train, x_test, y_train_t_min, y_test_t_min, y_train_t_max, y_test_t_max = train_test_split(x, y_t_min, y_t_max, test_size=0.2, random_state=42)
```

```python
t_min_regressor = RandomForestRegressor(random_state=42)
t_min_regressor.fit(x_train, y_train_t_min)

t_max_regressor = RandomForestRegressor(random_state=42)
t_max_regressor.fit(x_train, y_train_t_max)
```

```python
y_pred_t_min = t_min_regressor.predict(x_test)
y_pred_t_max = t_max_regressor.predict(x_test)
```
